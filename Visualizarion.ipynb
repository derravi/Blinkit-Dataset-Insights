```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('New_Datasets/New_Blinkit_Dataset.csv')
df.head(10)
```

```python
#Item type vs mrp

# df['Item_Type'].value_counts()

keys = df.groupby('Item_Type')['Item_MRP'].sum().keys()
value = df.groupby('Item_Type')['Item_MRP'].sum().values

plt.figure(figsize=(7,5))
plt.pie(value,labels=keys,autopct="%1.1f%%")
plt.title("Total Items Cell")
plt.show()
```

```python
#Item fat Content masuring

cat = df.groupby('Item_Fat_Content')['Item_Identifier'].count().keys()
val = df.groupby('Item_Fat_Content')['Item_Identifier'].count().values

plt.figure(figsize=(7,5))
plt.bar(cat,val,color=['skyblue','pink'])
plt.xlabel('Fat Category')
plt.ylabel('Total cell')
plt.title("Fat Categoty Cell")
plt.show()
```
